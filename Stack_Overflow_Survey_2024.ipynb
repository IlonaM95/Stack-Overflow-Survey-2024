---
## 1. How many respondents completed the survey?
---

```python
respondents_count = survey['ResponseId'].count()

print(respondents_count, 'respondents completed the survey')
```

---
## 2. How many respondents answered all the questions?---


```python
survey_dropna = survey.dropna()
notna_count = survey_dropna['ResponseId'].count()

print(notna_count, 'respondents answered all questions')
```

---
### Detailed verification

*How many questions were not answered by respondents and which questions were most frequently not answered?*

---

```python
survey_null_row = survey.isnull().sum(axis=1).sort_values(ascending = True)

print('How many questions were not answered by respondents?')
survey_null_row.describe()
```

```python
survey_null_col = survey.isnull().sum(axis=0).sort_values(ascending = False)

print('Total respondents who did not answer the individual questions:')
survey_null_col[survey_null_col > 0]
```

---
## 3. What are the values of the measures of central tendency for the respondents' work experience?
---


```python
exp_mean = survey['WorkExp'].mean()
exp_median = survey['WorkExp'].median()
exp_mode = survey['WorkExp'].mode()
```

```python
print("Values of measures of central tendency for respondents' work experience:")
print('Mean:', round(exp_mean,1))
print('Median:', exp_median)
print('Mode:', exp_mode[0])
```

```python
survey['RemoteWork'].unique()
```

```python
remote = survey['RemoteWork'][survey['RemoteWork']=='Remote'].count()

print(remote, 'respondents work remotely')
```

## 5. What percentage of respondents program in Python?
---

```python
schema[schema['question'].str.contains('language')]
```

```python
survey[survey.columns[survey.columns.str.contains('anguage')]].head()
```

```python
python_count = survey['LanguageHaveWorkedWith'][survey['LanguageHaveWorkedWith'].notna() & survey['LanguageHaveWorkedWith'].str.contains('Python')].count()
```

```python
python_pct = round(python_count / respondents_count * 100, 2)
```

```python
print(python_count, 'respondents program in Python, which represents',
      python_pct, '% of all respondents')
```

---
## 6. How many respondents learned programming through online courses?
---

```python
learn_code_online = survey['LearnCode'][survey['LearnCode'].notna() & survey['LearnCode'].str.contains('online')].count()
```

```python
learn_online_courses = survey['LearnCodeOnline'][(survey['LearnCodeOnline'].notna()
                                                 ) & (
                                                     survey['LearnCodeOnline'].str.contains('Courses')
                                                 )].count()
```

```python
print(learn_code_online, 'respondents learned programming online, of which',
      learn_online_courses, 'respondents learned through online courses')
```

---
## 7. What is the average and median salary in each country among respondents who program in Python?
---

```python
python_dev = survey[survey['LanguageHaveWorkedWith'].notna() & survey['LanguageHaveWorkedWith'].str.contains('Python')]
python_dev.head()
```

```python
python_grouped_country = python_dev.groupby(by = ['Country'])
```

```python
grouped_agg = python_grouped_country['ConvertedCompYearly'].agg(['median', 'mean']).sort_values(by = ['median', 'mean'],
                                                                                                ascending = [False, False])
grouped_agg['mean'] = round(grouped_agg['mean'], 2)
```

```python
print('Mean and median salary in each country among respondents who program in Python:')
grouped_agg.loc[grouped_agg['mean'].notna() & grouped_agg['median'].notna()]
```

---
## 8. What is the educational level of the 5 respondents with the highest salaries?
---

```python
comp_edu = survey.loc[:,['ConvertedCompYearly', 'EdLevel']]
top_comp_edu = comp_edu.loc[comp_edu['ConvertedCompYearly'].notna()].sort_values(by = ['ConvertedCompYearly'],
                                                                                ascending = False
                                                                               ).reset_index()
print('Education level of the 5 respondents with the highest salary:')
top_comp_edu.iloc[0:5]
```

---
## 9. What percentage of respondents program in Python, in each age group?
---

```python
grouped_age = python_dev.groupby(by = ['Age'])
```

```python
python_grouped_age = grouped_age[['ResponseId']].agg(lambda x: round( x.count() / respondents_count * 100, 2)
                                                    ).rename(columns = {'ResponseId': 'PythonPctRespondents'})
```

```python
print('Percentage of all respondents programming in Python by age group:')
python_grouped_age
```

---
## 10. What are the most common industries in which respondents who are in the 75th percentile of average salary and who work remotely are employed?
---

```python
remote_work = survey.loc[survey['RemoteWork']=='Remote']
```

```python
remote_third_q_comp = remote_work.loc[remote_work['ConvertedCompYearly'] >= remote_work['ConvertedCompYearly'].quantile(0.75)]
```

```python
groupby_industry = remote_third_q_comp.groupby(by = ['Industry'])
```

```python
industry_remote_third_q = groupby_industry[['ResponseId']].agg('count').sort_values(by = 'ResponseId',
                                                                                    ascending = False
                                                                                   ).rename(columns = {'ResponseId': 'Respondents_count'})
```

```python
print('Number of respondents by industry, working remotely and belonging to the top 25% of earners:')
industry_remote_third_q
```
